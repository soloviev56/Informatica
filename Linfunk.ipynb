## **Линейная функция**

```python
import numpy as np
from pylab import plot, show, legend

x_start=0; x_end=4
N_points=50
x=np.linspace(x_start,x_end,N_points)

def f(x,k,x0,y0):
    return(y0+k*(x-x0))

y1=f(x,1,0,0)
y2=f(x,1,0,1)
y3=f(x,1,0,2)
plot(x,y1,x,y2,x,y3,marker='x')
legend([1, 2, 3])
show()
```
